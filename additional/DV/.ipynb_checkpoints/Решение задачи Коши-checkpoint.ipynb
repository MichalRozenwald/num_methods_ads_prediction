## Решение задачи Коши для системы ОДУ
Используется метод Эйлера.
$$x_i=x_{i-1}+(t_i-t_{i-1})f(t_{i-1},x_{i-1},y_{i-1})$$

$$y_i=y_{i-1}+(t_i-t_{i-1})f(t_{i-1},x_{i-1},y_{i-1}),\quad i=1,2,3,\dots,n.$$
Порядок аппроксимации метода - O(h)

```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import numpy as np
from math import sin, cos, sqrt, log, e
```

```python
class TabulatedFunction:

    def __init__(self, tabulated_values, grid):
        self.values = tabulated_values
        self.grid = grid

    def __getitem__(self, i):
        return self.values[i]
```

```python
def diff_equastions(func1, func2, x_0, y_0, grid):
    x = [x_0,]
    y = [y_0,]
    step = grid[1] - grid[0]
    for i in range(1, len(grid)):
        x_old = x[i - 1]
        y_old = y[i - 1]
        t_old = grid[i - 1]
        x_new = x_old + step * func1(x_old, y_old)
        y_new = y_old + step * func2(x_old, y_old)
        x.append(x_new)
        y.append(y_new)
    return TabulatedFunction(np.array(x), grid), TabulatedFunction(np.array(y), grid)
```

Решим систему дифференциальных уравнений:
$$ \left\{\begin{array}{l} {\frac{dx}{dt} =-2x+4y,} \\ {\frac{dy}{dt} =-x+3y,} \end{array}\right. $$

$$ x\left(0\right)=3,y\left(0\right)=0$$

Аналитическое решение системы:
$$ \left\{\begin{array}{l} {x\left(t\right)=4e^{-t} -e^{2t},} \\ {y\left(t\right)=e^{-t} -e^{2t}} \end{array}\right. $$

```python
dx = lambda x, y: -2*x + 4*y
dy = lambda x, y: -x + 3*y
x_0 = 3
y_0 = 0
grid = np.linspace(0, 1, 100)
```

```python
x_precise = lambda t: 4*e**(-t) - e**(2*t)
y_precise = lambda t: e**(-t) - e**(2*t)
x_vals = [x_precise(t) for t in grid]
y_vals = [y_precise(t) for t in grid]
```

```python
x, y = diff_equastions(dx, dy, x_0, y_0, grid)
```

```python
fig, axis = plt.subplots(nrows=1, ncols=2, figsize=(20,10))

axis[0].plot(x.grid, x.values, label="approx")
axis[0].set_xlabel("t")
axis[0].set_ylabel("x")
axis[0].set_title("x(t)")
axis[0].plot(grid, x_vals, label="precise")
axis[0].set_xlabel("t")
axis[0].set_ylabel("x")
axis[0].set_title("x(t)")
axis[0].legend()

axis[1].plot(y.grid, y.values, label="approx")
axis[1].set_xlabel("t")
axis[1].set_ylabel("y")
axis[1].set_title("y(t)")
axis[1].plot(grid, y_vals, label="precise")
axis[1].set_xlabel("t")
axis[1].set_ylabel("y")
axis[1].set_title("y(t)")
axis[1].legend()
```

```python
def error(x_approx, x_precise):
    return np.max(np.abs(np.array(x_approx) - np.array(x_precise)))    
```

```python
nnodes = np.linspace(10, 1000, 100)
steps = []
errors_x = []
errors_y = []
for n in nnodes:
    grid = np.linspace(0, 1, n)
    steps.append(grid[1]-grid[0])
    x_prec_vals = [x_precise(t) for t in grid]
    y_prec_vals = [y_precise(t) for t in grid]
    x, y = diff_equastions(dx, dy, x_0, y_0, grid)
    errors_x.append(error(x.values, x_prec_vals))
    errors_y.append(error(y.values, y_prec_vals))
plt.figure(figsize=(8,5))
plt.plot(nnodes, errors_x, label="errors x")
plt.xlabel("N nodes")
plt.ylabel("error")
plt.plot(nnodes, errors_y, label="errors y")
plt.legend()
```

```python
plt.figure(figsize=(8,5))
plt.plot(nnodes, errors_x, label="error x")
plt.plot(nnodes, errors_y, label="error y")
plt.xlabel("log(N nodes)")
plt.ylabel("log(error x)")
plt.gca().set_xscale('log')
plt.gca().set_yscale('log')
plt.legend(loc=3)
```

```python
coefs = np.polyfit(np.log(nnodes), np.log(errors_x), deg=1)
```
