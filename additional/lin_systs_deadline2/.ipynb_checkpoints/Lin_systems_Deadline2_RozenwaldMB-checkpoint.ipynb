# Решение системы линейных алгебраических уравнений
## Метод Крамера

$Ax = b$ -- Система линейных уравнений

$\Delta$ -- определитель матрицы А

$\Delta_i$ -- определитель матрицы системы, в которой столбец b заменяет $i$-й столбц матрицы А

$ x_i = \frac{\Delta_i}{\Delta}$ -- Решение системы находится с помощью формулы Крамера, когда определитель матрицы А не вырожден



###  Условия сходимости и применимости
Если $\Delta \neq 0$, то система совместна и имеет единственное решение.
При этом, если определитель системы равен нулю, то система может быть либо совместной (иметь либо бесконечно много решений), либо несовместной (не имеет решений).

Метод Крамера применяется для решения систем линейных алгебраических уравнений, в которых число неизвестных переменных равно числу уравнений и определитель основной матрицы А отличен от нуля. Т.е. система имеет единственное решение.


### Корректность
Пусть $\Delta \neq 0$.

Предположим, что единственное решение системы - $x_1, \dots, x_n$. 

Домножим $x_j$ на $j$-ый столбец $A$. Тогда определитель $\Delta$ станет равным $\Delta * x_j$. 

Добавим $j$-ому столбцу линейную комбинацию остальных столбцов, перемноженных с $x_1, \dots, x_{j-1}, x_{j+1}, \dots, x_n$. При этом преобразовании определитель не изменится.

Тогда в $i$-ой строке матрицы будет стоять:
$a_{i1}, a_{i2}, \dots, a_{i1}x_1 + \dots + a_{ij}x_j + \dots + a_{in}x_n, \dots, a_{in}$
Это значит, что в $j$-ом столбце будет стоять столбец b.

Получаем, что

$ x_i\Delta = \Delta_i$  => $ x_i = \frac{\Delta_i}{\Delta}$

Ч.т.д.

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline 
import math
```

```python
def Det(A):
    return np.linalg.det(A)
```

```python
def Kramer(A, b):
    det = Det(A)
    
    A_size = A.shape[1]
    x = [0]*A_size

    if (det != 0):
        for i in range(A_size):
            A_i = A.copy()
            A_i[:, i] = b
            det_i = float(Det(A_i))
            x_i = det_i / det
            x[i] = x_i
        return x
    
    return "det == 0, NO SOLUTION"
```

```python
A = np.array(([1,2,3],[2.0001, 3.999, 6], [15,3,6]))
b = np.array([1, 10, 3])

Kramer(A, b)
```

### Норма матрицы $||A||_{\infty} = \max_{1\leq i \leq n} \sum_{j=1}^n |a_{ij}|$

```python
def Norm(matrix):
    return np.max(np.sum(np.abs(matrix), axis=1))
```

### Числа обусловленности  $\mu (A)=||A||\cdot ||A^{{-1}}||$

```python
def Condition_number(A):
    A_inv = np.linalg.inv(A)
    return Norm(A) * Norm(A_inv)
```

```python
def Matrix_det_morm_cond(A):
    print "det: ", Det(A)
    print "norm: ", Norm(A)
    print "Condition_number: ", Condition_number(A)
```

### Hевязки решения

```python
def Solve_residual(A, b):
    x = Kramer(A, b)
    print "x =", x
    resid = np.abs(b - A.dot(x))
    print "residuals: |b - Ax| =", resid
    resid_norm = np.linalg.norm(resid)
    print "residual norm: ", resid_norm

    return x, resid, resid_norm
```

# Интерполяции кубическими сплайнами дефекта 1


Некоторая функция f(x) задана на отрезке <math>[a,b]</math>, разбитом на части $[x_{i-1},x_i], a=x_0< x_1< ... <x_N=b$. 

**Кубическим сплайном дефекта** 1 называется функция <math>S(x)</math>, которая:
* на каждом отрезке $[x_{i-1},x_i]$ является многочленом степени не выше третьей;
* имеет непрерывные первую и вторую производные на всём отрезке <math>[a,b]</math>;
* в точках <math>$x_i$</math> выполняется равенство $S(x_i) = f(x_i)$, т. е. сплайн <math>S(x)</math> интерполирует функцию f в точках $x_i$.

Для однозначного задания сплайна перечисленных условий недостаточно, для построения сплайна необходимо наложить дополнительные требования.

Естественным кубическим сплайном называется кубический сплайн, удовлетворяющий также граничным условиям вида: <math>S''(a) = S''(b) = 0.</math>


### Метод нахождения коэффициентов сплайнов 
Основан на том, что на кажом отрезке $[x_{i−1} , x_i]$ интерполируем функцию полиномом степени не выше третьей, у которого первая и вторая производные непрерывны на всём отрезке. 

$S_i$ - i-ый сплайн

В точках $x_i$ выполняется равенство $S(x_i) = f (x_i)$.

На кажом отрезке разбиения $[x_{i−1} , x_i ]$ сплайн задается таким соотношением:

$ S_i(x) = m_i * (x_{i+1} − x)^3 + m_(i+1) (x − x_i)^3 + A_i x + B_i$

m ("x_kram") – вектор значений вторых производных. 

Находим m с 1 по n-1, из решения СЛАУ такого вида:

Краевые условия на m_0 , m_n , как правило их принимают равными 0.

$A * x = b$ , где А - трехдиагональная, такая что:

на главной диагонали стоят суммы соответствующих значений шага деленные на 3, а слева и сверху находится значение соответствующий точки деленное на 6.

$ A =  \begin{bmatrix}
    (δ_0 +δ_1)/3      & δ_1/6  &0  & \dots & 0 \\
    δ_1/6        &(δ_1 +δ_2)/3  & 0 & \dots &  0 \\
    \dots & & & & δ_{i-2}/6 \\
    0        & \dots & δ_{i-2}/6 & & (δ_{i-1} +δ_{i-2})/3
\end{bmatrix}
$

$b = \begin{bmatrix}
    (f_2 −f_1) / δ1 - (f_1 −f_0) / δ_0 \\
    \dots \\
        (f_{i+1} −f_i) / δ_i - (f_i −f_{i-1}) / δ_{i-1}\\
\end{bmatrix}
$

Решив СЛАУ, выражаются коэффициенты A("d_1"), B("d_2"):
$ A_i  = (f_{i+1} −f_k) / δ_i - δ_i / 6 * (m_{i+1} - m_i) $

$ B_i = f_i − 1/6 * m_i δ^2 − A_i * x_i $



```python
def Interpolate_steps_tabu(x_steps, f_tabu):
    x_dif = [x_steps[i] - x_steps[i-1] for i in range(1, len(x_steps))]
    b = np.array([(f_tabu[i+1]-f_tabu[i])/x_dif[i] - (f_tabu[i]-f_tabu[i-1])/x_dif[i-1] for i in range(1, len(x_steps) - 1)])

    n = len(x_steps) - 1
    A = np.zeros((n - 1, n - 1))
    for i in range(n-2):
        A[i][i] = (x_dif[i] + x_dif[i+1]) / 3
        A[i+1][i] = x_dif[i+1] / 6
        A[i][i+1] = x_dif[i+1] / 6
    A[n-2][n-2] = (x_dif[n-2]+x_dif[n-1]) / 3
    
    x_kram = [0] +  Kramer(A, b) + [0]
    d_1 = [(f_tabu[i + 1] - f_tabu[i])/x_dif[i] - x_dif[i]*(x_kram[i+1]-x_kram[i])/6 for i in range(n)]
    d_2 = [f_tabu[i] - x_kram[i]*x_dif[i]*x_dif[i]/6 - d_1[i]*x_steps[i] for i in range(n)]
    splines = []
    sp_coef_cur = [0,0,0,0]
    for i in range(n):
        sp_coef_cur[0] = 1/(6*x_dif[i])*(x_kram[i+1] - x_kram[i])
        sp_coef_cur[1] = 1/(2*x_dif[i])*(x_kram[i]*x_steps[i+1] - x_steps[i]*x_kram[i+1])
        sp_coef_cur[2] = d_1[i] + 1/(6*x_dif[i])*(3*x_steps[i]*x_steps[i]*x_kram[i+1] - 3*x_steps[i+1]*x_steps[i+1]*x_kram[i])
        sp_coef_cur[3] = d_2[i] + 1/(6*x_dif[i])*(x_kram[i]*pow(x_steps[i + 1],3) - x_kram[i+1]*pow(x_steps[i], 3))
        splines.append(sp_coef_cur) 
        sp_coef_cur = [0,0,0,0]
        
    f_tabu_interp = []
    x_steps_all = []
    for i in range(n):
        x_st = np.linspace(x_steps[i], x_steps[i+1], num=163)
        spline = splines[i]
        f = list(map(lambda x: spline[3] + x*spline[2] + x*x*spline[1] + x*x*x*spline[0], x_st))
        x_steps_all += list(x_st)
        f_tabu_interp += f
                                
    return f_tabu_interp, x_steps_all
```

```python
def plot_difs(x_steps_closer, f_tabu_closer, x_steps_all, f_tabu_interp, er, printErrorPlot=False):
    plt.title(u"Interpolation S(x) vs real f(x)")
    plt.xlabel(u'x')
    plt.ylabel(u'y')
    plt.plot(x_steps_all, f_tabu_interp, label = u'S_3_1(x)')
    plt.plot(x_steps_closer, f_tabu_closer, label = u'f(x)')
    plt.legend()
    plt.show()
    
    if (printErrorPlot):
        plt.title(u"Error: f(x) - S(x)")
        plt.xlabel(u'x')
        plt.ylabel(u'f(x) - S(x)')
        plt.plot(x_steps_all, er)
        plt.show()
```

```python
def Interpolate_compare(func, a, b, steps_num, printErrorPlot=False):
    x_steps = np.linspace(a, b, num=steps_num)
    
    func_lambda = func.lambda_func()
    f_tabu = list(map(lambda x:func_lambda(x),x_steps))
    f_tabu_interp, x_steps_all = Interpolate_steps_tabu(x_steps, f_tabu)
    
    x_steps_closer = x_steps_all
    f_tabu_closer = list(map(lambda x:func_lambda(x), x_steps_closer))
    er = np.array(f_tabu_closer) - np.array(f_tabu_interp)
    
    x_steps_closer =  np.linspace(a, b, num=len(x_steps_all)*5)
    f_tabu_closer = list(map(lambda x:func_lambda(x), x_steps_closer))
    

    plot_difs(x_steps_closer, f_tabu_closer, x_steps_all, f_tabu_interp, er, printErrorPlot)
```

### Класс функций полином. На примере полиномов будем рассматривать поведение гладких функций. 

```python
class Polynomial(object):
    def __init__(self, coefficients):
        self.coeff = coefficients
        while(len(self.coeff) < 7):
            self.coeff.append(0)

    def __call__(self, x):
        # Return polinomial value
        s = 0
        for i in range(len(self.coeff)):
            s += int(self.coeff[i])*x**i
        return s

    def PolyPrint(self):
        if (len(self.coeff) == 0):
            return ('Empty function')
        poly_string = ''
        for i in range(len(self.coeff) - 1, 0, -1):
            if (self.coeff[i] != 0):
                poly_string += str(self.coeff[i]) + 'x^' + str(i) +' + '
        poly_string += str(self.coeff[0])

        return (poly_string)

    def __mul__(self, num):
        result_coeff = list(map(int, self.coeff[:])) # copy!
        for i in range(len(self.coeff)):
            result_coeff[i] *= num
        return Polynomial(result_coeff)


    def __add__(self, other):
        # Return self + other as Polynomial object
        # Start with the longest list and add in the other
        if len(self.coeff) > len(other.coeff):
            result_coeff = list(map(int, self.coeff[:]))  # copy
            for i in range(len(other.coeff)):
                result_coeff[i] += int(other.coeff[i])
        else:
            result_coeff = list(map(int, other.coeff[:])) # copy
            for i in range(len(self.coeff)):
                result_coeff[i] += int(self.coeff[i])
        return Polynomial(result_coeff)
    
    def differ_max(self):
        # only for polinoms with degree <= 2
        x_max = - float(self.coeff[1]) / (2*self.coeff[0])
        return self.coeff[1] + 2 * self.coeff[1] * x_max
    
    def lambda_func(self):
        # only for polinoms with degree <= 2
        #if (self.coeff.size() = 2) { self.coeff}
        return lambda x: self.coeff[0] + x * self.coeff[1] + x * x * self.coeff[2] + x * x * x * self.coeff[3] + (x**4)*self.coeff[4] + (x**5)*self.coeff[5]
    
    def lambda_func_ans(self, x):
        # only for polinoms with degree <= 2
        return self.coeff[0] + x * self.coeff[1]
```

```python
poly = Polynomial([1,2,3,4,5,6])
print poly.PolyPrint()
print 'poly(0) = ', poly(0)
print 'poly(1) = ',poly(1)

Interpolate_compare(poly, -10, 10, 10, True)
```

```python
Interpolate_compare(poly, -10, 10, 100, True)
```

С увеличением числа узлов в сетке квадратичный многочлен приближается с помощью кубических сплайнов довольно точно (ошибка равна нулю везде, кроме краев). Это проиходит потому, что полином второй степени хорошо приближается полиномом третьей степени. А на краях обобщение работает с ошибкой из-за ограничивающих краевых условий.

### Разрывные функции будем задавать кусочно, как два разлиных полинома, заданных на интервалах poly_a на (-inf; x_0) и poly_b на [x_0; inf).

$f(x) = \begin{cases} a_1 x + a_0 & x < x_0 \\ b_1 x + b_0 & x \geq x_0 \end{cases}$

```python
class Discontinuous(object):
    # poly_a with coeffs (a_1, a_0) when x in  (-inf; x_0), esle - poly_b with coeffs (b_1, b_0)
    def __init__(self, coefficients_a, coefficients_b, x_0):
        self.coeff_a = coefficients_a
        self.coeff_b = coefficients_b
        self.poly_a = Polynomial(coefficients_a)
        self.poly_b = Polynomial(coefficients_b)
        self.x_0 = x_0

    def __call__(self, x):
        # Return polinomial value
        if (x < self.x_0):
            s = self.poly_a(x)
        else:
            s = self.poly_b(x)
        return s
    
    def DiscontinPrint(self):
        disc_string = self.poly_a.PolyPrint() + " , x < " + str(self.x_0) + '\n' + self.poly_b.PolyPrint() + " , x > " + str(self.x_0) 
        return (disc_string)
    
    def differ_max(self):
        return max( self.poly_a.differ_max(), self.poly_b.differ_max())
    
    def lambda_func(self):
        return lambda x: (self.poly_a.coeff[0] + x * self.poly_a.coeff[0]) if x < self.x_0 else (self.poly_b.coeff[0] + x * self.poly_b.coeff[0])
    
```

```python
disc = Discontinuous([1, 2], [2, 3], 0)
print 'disc function : \n', disc.DiscontinPrint()
print 'disc(0) = ', disc(0)
print 'disc(-0.1) = ', disc(-0.1)
print 'disc(0.1) = ', disc(0.1)


func = disc
Interpolate_compare(func, -10, 10, 10)
```

```python
Interpolate_compare(func, -10, 10, 100, True)
```

На более густой сетке ошибка приближения разрывной функции с помощью сплайнов равна 0 везде, кроме области разрыва. На краях обобщение работает без ошибкой из-за того, что функции простые - линейные.
А в области развыва даже с увеличением числа узлов в сетке ошибка остается достаточно большой.

Применяя метод интерполяции, использующий кубический сплайн дефекта 1, для каждой из рассмотренных типов функций было показано, что с увеличением шага сетки точность приближения функции растёт и ошибка значительно уменьшается и почти на всей области стремится к нулю.
