## Метод Крамера решения систем линейных уравнений
#### Описание алгоритма
Система линейных уравнений:
$$Ax = F$$
Если определитель матрицы системы A не равен нулю, то система совместна и имеет единственное решение, которое находится по формулам Крамера:

$$ x_i = \frac{\Delta_i}{\Delta}$$

где $\Delta$ - определитель матрицы системы, $\Delta_i$ - определитель матрицы системы, где вместо $i$-го столбца стоит столбец F правых частей.

#### Обоснование корректности

Если $\Delta \neq 0$, то система совместна и имеет единственное решение.

Допустим, $\Delta \neq 0$. Предположим, что $x_1, \dots, x_n$ - единственное решение системы. Домножим $j$-ый столбец $A$ на $x_j$. Тогда определитель $\Delta$ также умножится на $x_j$. Прибавим к $j$-ому столбцу линейную комбинацию остальных столбцов, перемноженных с $x_1, \dots, x_{j-1}, x_{j+1}, \dots, x_n$. При этом преобразовании определитель не изменится.

Тогда в $i$-ой строке матрицы будет стоять:
$$a_{i1}, a_{i2}, \dots, a_{i1}x_1 + \dots + a_{ij}x_j + \dots + a_{in}x_n, \dots, a_{in}$$
Но это значит, что в $j$-ом столбце будет стоять столбец F.
Отсюда, $$ x_i\Delta = \Delta_i$$ 
$$ x_i = \frac{\Delta_i}{\Delta}$$
#### Условия сходимости и применимости
Метод Крамера применим только если определитель матрицы A не равен 0. При этом, если определитель системы равен нулю, то система может быть как совместной, так и несовместной, то есть иметь либо бесконечно много решений, либо ни одного.

Соответственно метод Крамера применим, только если система имеет единственное решение.

```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
# A: np.array((n, m))
# B: np.array((n, 1))
def kramer(A, F):
    det = np.linalg.det(A)
    if not det:
        raise RuntimeError("Determinant equals to 0.")
    roots = []
    for x in range(A.shape[1]):
        tmp = A.copy()
        tmp[:, x] = F
        roots.append(float(np.linalg.det(tmp)) / det)
    return roots
```

Норма матрицы
$$||A||_{\infty} = \max_{1\leq i \leq n} \sum_{j=1}^n |a_{ij}|$$

```python
def matrixNorm(matrix):
    return np.max(np.sum(np.abs(matrix), axis=1))
```

$$ A = \begin{bmatrix}
    1 & 2 & 3 \\
    2.0001 & 3.999 & 6 \\
    15 & 3 & 6
\end{bmatrix},
B = \begin{bmatrix}
    1 & \frac{1}{2} & \frac{1}{3} \\
    \frac{1}{2} & \frac{1}{3} & \frac{1}{4} \\
    \frac{1}{3} & \frac{1}{4} & \frac{1}{6}
\end{bmatrix},
C = \begin{bmatrix}
    10^{6} & 2 \\
    10^{13} & 2 \\
\end{bmatrix}$$    

```python
A = np.array(([1,2,3],[2.0001, 3.999, 6], [15,3,6]))
B = np.array(([1, 0.5, float(1)/3],
              [0.5, float(1)/3, 0.25],
              [float(1)/3, 0.25, float(1)/6]))
C = np.array(([10**6, 2], [10**13, 2]))
print "Определители матриц:"
print "||A|| =", np.linalg.det(A)
print "||B|| =", np.linalg.det(B)
print "||C|| =", np.linalg.det(C)

print "\nМатричные нормы:"
print "||A|| =", matrixNorm(A)
print "||B|| =", matrixNorm(B)
print "||C|| =", matrixNorm(C)

print "\nЧисла обусловленности:"
print "Cond(A) =", matrixNorm(A)*matrixNorm(np.linalg.inv(A))
print "Cond(B) =", matrixNorm(B)*matrixNorm(np.linalg.inv(B))
print "Cond(C) =", matrixNorm(C)*matrixNorm(np.linalg.inv(C))
```

```python
F = np.array([1, 10, 3])
resA = kramer(A, F)
print "x =", resA
print "|F - Ax| =", np.abs(F - A.dot(resA))
```

```python
F = np.array([20, 10, 3])
resB = kramer(B, F)
print "x =", resB
print "|F - Bx| =", np.abs(F - B.dot(resB))
```

```python
F = [5, 10]
resC = kramer(C, F)
print "x =", resC
print "|F - Cx| =", np.abs(F - C.dot(resC))
```

## Модуль интерполяции кубическими сплайнами дефекта 1
#### Описание метода
Функция $f(x)$ задана на отрезке $[a,b]$, разбитым по сетке $\{x_{i}\}_{i=1}^N$.  Кубическим сплайном дефекта 1 называется функция $S(x)$, которая
- на каждом отрезке $[x_i, x_{i+1}]$ является многочленом 3 степени
- имеет непрерывные первую и вторую производные на всём отрезке $[a,b]$
- $\forall x_i: S(x_i)=f(x_i)$
- $S''(a) = S''(b) = 0$

Сплайн на отрезке $[x_i, x_{i+1}]$ представим в виде:
$$S_i(x) = a_i + b_i(x - x_i) + {c_i\over2}(x-x_i)^2 + {d_i\over6}(x - x_i)^3$$
$$S_i\left(x_i\right) = a_i, \quad S'_i(x_i) = b_i, \quad S''_i(x_i) = c_i$$
Условия непрерывности всех производных:
$$S_i\left(x_{i-1}\right) = S_{i-1}(x_{i-1})$$
$$S'_i\left(x_{i-1}\right) = S'_{i-1}(x_{i-1})$$
$$S''_i\left(x_{i-1}\right) = S''_{i-1}(x_{i-1})$$
Условие интерполяции:
$$S_i\left(x_{i}\right) = f(x_{i})$$
Обозначим $\quad h_i = x_i - x_{i-1}, \quad f_{i} = f(x_{i})$

Отсюда получаем формулы для вычисления коэффициентов сплайна:
$$h_{i}c_{i} + 2(h_i + h_{i+1})c_{i+1} + h_{i+1}c_{i+2} =
6\left({{f_{i+2} - f_{i+1}}\over{h_{i+1}}} - {{f_{i+1} - f_{i}}\over{h_{i}}}\right)$$
Зная, что $c_1 = c_n = 0$, надо решить систему из N-2 линейных уравнений на $c_2, \dots, c_{n-1}$
Через $c_i, c_{i+1}$ выражаются остальные коэффициенты сплайнов.
$$a_i = f\left(x_{i}\right)$$
$$b_i = {{f_{i+1} - f_{i}}\over{h_i}} + {{h_i(2c_{i+1} + c_{i})}\over6}$$
$$d_i = {{c_{i+1} - c_{i}}\over{h_i}}$$


```python
class TabulatedFunction:
    
    def __init__(self, tabulated_values, grid):
        self.values = tabulated_values
        self.grid = grid
        
    def __getitem__(self, i):
        return self.values[i]

        
class Interpolation:
    
    def __init__(self, tabulated_function, splines):
        self.grid = tabulated_function.grid
        self.splines = splines
        
    def __call__(self, x):
        if (x < self.grid[0]) or (x > self.grid[len(self.grid) - 1]):
            raise RuntimeError("x is not in domain of interpolation") 
        for i in range(len(self.grid) - 1):
            if (x >= self.grid[i]) & (x <= self.grid[i+1]):
                break
        return self.splines[i](x)
        
class Polinomial:
    def __init__(self, degree, coefs, x0):
        self.degree = degree
        self.coefs = coefs
        self.x0 = x0
        if degree + 1 != len(coefs):
            raise RuntimeError("Polinom is initialised incorrectly")
            
    def __call__(self, x):
        res = 0
        for i in range(self.degree + 1):
            res += self.coefs[i] * (x - self.x0)**i
        return res
        
def interpolate(tabulated_function):
    
    grid = tabulated_function.grid
    N = len(grid) - 2
    matrix = np.zeros((N, N))
    values = np.zeros(N)
    for i in range(1, N+1):
        
        h_0 = grid[i] - grid[i-1]
        h_1 = grid[i+1] - grid[i]
        
        if i != 1:
            matrix[i-1, i-2] = h_0
        if i != N:
            matrix[i-1, i] = h_1
        
        matrix[i-1, i-1] = 2*(h_0 + h_1)
        delta_1 = float(tabulated_function[i+1] - tabulated_function[i])
        delta_0 = float(tabulated_function[i] - tabulated_function[i-1])
        values[i-1] = 6 * (delta_1 / h_1 - delta_0 / h_0)
    derivatives_2 = np.linalg.solve(matrix, values)

    splines = []     
    for i in range(N+1):
        der2_0 = float(derivatives_2[i-1] if i != 0 else 0)
        der2_1 = float(derivatives_2[i] if i != N else 0)
        
        h = grid[i+1] - grid[i]
        der3 = (der2_1 - der2_0) / h
        
        f0 = float(tabulated_function[i])
        f1 = float(tabulated_function[i+1])
        
        der1 = (f1 - f0) / h - der2_1 * h / 6 - der2_0 * h / 3
        
        p = Polinomial(3, [f0, der1, der2_0 / 2, der3 / 6], grid[i])
        splines.append(p)
    return Interpolation(tabulated_function, splines)
                             
```

```python
def tabulateFunction(function, interval, nnodes):
    grid = np.linspace(interval[0], interval[1], nnodes)
    tabulated = [function(x) for x in grid]
    return TabulatedFunction(tabulated, grid)
```

### Пример гладкой функции.
$$f(x) = -x^{6}$$
$$x \in [-10, 10]$$
Выбереем две равномерные сетки на отрезке $[-10, 10]$ - из 10 узлов и из 1000 узлов.

```python
func = lambda x: x**6
interval = [-100, 100]
nnodes = 10
tab_func = tabulateFunction(func, interval, nnodes)
interpolation = interpolate(tab_func)
ys = [interpolation(x) for x in tab_func.grid]
```

```python
print "Сетка из 10 узлов"
fig, axis = plt.subplots(nrows=1, ncols=3, figsize=(20, 7))

axis[0].plot(tab_func.grid, ys)
axis[0].set_title("Interpolation")

axis[1].plot(tabulateFunction(func, interval, 1000).grid, tabulateFunction(func, interval, 1000).values)
axis[1].set_title("Ground truth")

axis[2].plot(tab_func.grid, np.array(ys) - np.array(tab_func.values))
axis[2].set_title("Error")

plt.show()
```

### Пример разрывной функции 
возьмем $$f(x) = \begin{cases} \sin(x) & x < 0 \\ \cos(x) & x \geq 0 \end{cases}$$

$$f^{(4)}(x) = \begin{cases} \sin(x) & x < 0 \\ \cos(x) & x \geq 0 \end{cases}$$

на отрезке $[-5, 5]$

```python
func = lambda x: np.sin(x) if x < 0 else np.cos(x)
interval = [-5, 5]
nnodes = 10
tab_func = tabulateFunction(func, interval, nnodes)
interpolation = interpolate(tab_func)
ys = [interpolation(x) for x in tab_func.grid]
print "Сетка из 10 узлов"
fig, axis = plt.subplots(nrows=1, ncols=3, figsize=(20, 7))

axis[0].plot(tab_func.grid, ys)
axis[0].set_title("Interpolation")

axis[1].plot(tabulateFunction(func, interval, 1000).grid, tabulateFunction(func, interval, 1000).values)
axis[1].set_title("Ground truth")

axis[2].plot(tab_func.grid, np.array(ys) - tab_func.values)
axis[2].set_title("Error")
```
